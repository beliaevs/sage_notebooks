```python
nindex = 2
```

```python
def inPoint(h):
    return (sqrt(1-h*h), h)
```

```python
def rotate(p, angle):
    return (p[0]*cos(angle)-p[1]*sin(angle), p[0]*sin(angle) + p[1]*cos(angle))
```

```python
def sym(p):
    return (-p[0], p[1])
```

```python
def inAngle(h):
    return asin(h)-asin(h/nindex)
```

```python
def outPoint(h):
    a = inAngle(h)
    return rotate(sym(rotate(inPoint(h), -a)), a)
```

```python
circle((0, 0), 1) + sum([line([(2, h), inPoint(h), outPoint(h), lPoint(h)], color='red', thickness=0.25) for h in [k*0.05 for k in range(-9, 10)] ])
```

```python
def lPoint(h):
    p = outPoint(h)
    lBorder = -3
    return (lBorder, tan(2*inAngle(h))*(lBorder-p[0]) + p[1]) 
```
